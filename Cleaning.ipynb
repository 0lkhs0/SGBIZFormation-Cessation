# Cleanning of Dataset

1. Getting rid irrelevant header information and footnotes
2. Change dtype of numerical values from type objects into int
3. Change dtype of date column into from type object into datetime
4. Making a dataset specific to the time period from when covid started to present

Dataset used: 
1. Formation Of Business Entities, By Industry, Monthly
2. Cessation Of Business Entities, By Industry, Monthly

```python
# Import libraries

import pandas as pd
# import seaborn as sb
# import numpy as np
# import matplotlib.pyplot as plt
```

```python
# Reading the csv files

formation = pd.read_csv('FormationData.csv')

cessation = pd.read_csv('CessationData.csv')
```

```python
# removing irelevalent header information and footnotes
# formation dataset

holder = formation.drop(formation.index[0:9], axis=0)
formation = holder.drop(formation.index[26:49], axis=0)
```

```python
# Orientating the dataset and setting the header
formation = formation.T

formation = formation.reset_index(drop=True)

header = formation.iloc[0]

formation = formation[1:]

formation.columns = header

#formation
```

```python
#removing whitespaces from the column headers

formation.columns = formation.columns.str.strip()

print(formation.columns)
```

```python
# Changing the column (industry) values into numeric

formation.iloc[:, 1:] = formation.iloc[:, 1:].apply(pd.to_numeric)
```

```python
# Changing the colume (Data Series) value into date time

formation['Data Series'] = formation['Data Series'].str.strip()
formation['Data Series'] = pd.to_datetime(formation['Data Series'], format='%Y %b')
```

```python
# Checking conversion
formation['Data Series']
```

```python
#Save dataset 
formation.to_csv('Formation_Cleaned.csv', index=False)
```

```python
# Filter time period from when covid-19 first started 
covid_formation = formation[formation['Data Series'] >= "2020-01-01"]
# covid_formation.reset_index(inplace=True)

# Save the dataset

covid_formation.to_csv('Covid_Formation.csv', index=False)
```

```python
#Create dataset for pre-covid

pre_covid_formation = formation[formation['Data Series'] < "2020-01-01"]
# pre_covid_formation.reset_index(inplace=True)

# Save the dataset
pre_covid_formation.to_csv('Pre_covid_Formation.csv', index=False)
```
